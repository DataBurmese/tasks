```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("employee_performance.csv")

# Ensure the 'Date' column is a datetime object
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Calculate the average performance score for each department
average_performance_by_department = df.groupby('Department')['Performance_Score'].mean()

# Plot the bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x=average_performance_by_department.index, y=average_performance_by_department.values, palette='coolwarm')

# Labels and title
plt.xlabel("Department")
plt.ylabel("Average Performance Score")
plt.title("Average Performance Score by Department")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Calculate total tasks completed by department
total_tasks_by_department = df.groupby('Department')['Tasks_Completed'].sum()

# Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(total_tasks_by_department, labels=total_tasks_by_department.index, autopct='%1.1f%%', colors=sns.color_palette("Set2"), startangle=140)

# Title
plt.title("Proportion of Total Tasks Completed by Department")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure the 'Date' column is a datetime object
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract Year, Month, and Hour from the 'Date' column
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Hour'] = df['Date'].dt.hour

# Calculate average tasks completed per hour across months
tasks_per_hour_monthly = df.groupby(['Year', 'Month', 'Hour'])['Tasks_Completed'].mean().reset_index()

# Pivot to get a better structure for plotting
tasks_pivot = tasks_per_hour_monthly.pivot(index='Hour', columns=['Year', 'Month'], values='Tasks_Completed')

# Plot the line graph
plt.figure(figsize=(10, 5))
tasks_pivot.plot(marker='o', figsize=(10, 5), colormap='viridis')

# Labels and title
plt.xlabel("Hour of the Day")
plt.ylabel("Average Tasks Completed")
plt.title("Trends in Average Tasks Completed Per Hour Across Months")
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
